```python
%matplotlib inline

import pandas as pd
import requests, json

deaths_url = "https://onemocneni-aktualne.mzcr.cz/api/v2/covid-19/umrti.json"

deaths_json = requests.get(deaths_url).json()['data']

deaths_df = pd.read_json(json.dumps(deaths_json))
```

```python
bins = [20+(x*10)for x in range(0,10)]
ax = deaths_df.plot.hist(bins=bins, alpha=0.5)
```

```python
bins = [20+(x*5)for x in range(0,20)]
ax = deaths_df.plot.hist(bins=bins, alpha=0.5)
```
